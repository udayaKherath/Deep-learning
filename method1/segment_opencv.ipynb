```python
import numpy as np
from skimage import io, measure
import matplotlib.pyplot as plt
import cv2
from skimage import io, img_as_ubyte
import imageio
```

```python
img = imageio.imread('/content/drive/My Drive/Colab Notebooks/datasets/images/10.jpg')
plt.imshow(img)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
mask = cv2.inRange(hsv, (54, 88, 100), (150, 250, 250))

plt.imshow(mask)
```

```python
from scipy import ndimage as nd
closed_mask = nd.binary_closing(mask, np.ones((50, 50)))
plt.imshow(closed_mask)
```

```python
plt.subplot(1, 2, 1)
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.imshow(closed_mask)
```
